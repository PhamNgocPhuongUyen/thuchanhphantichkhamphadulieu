```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
try:
    wine_data = pd.read_csv("winequality-red.csv")
    print("Dữ liệu đã được tải thành công!")
except FileNotFoundError:
    print("Lỗi: Không tìm thấy tệp 'winequality-red.csv'.")
    print("Vui lòng đảm bảo tệp nằm trong cùng thư mục hoặc cung cấp đường dẫn chính xác.")
    exit() 
```

NHIỆM VỤ 1 

```python
wine_data.head(5)
wine_data.dtypes
wine_data.shape
```

```python
print("\n--- Thống kê mô tả tổng quan (.describe()) ---")
print(wine_data.describe())
```

```python
column_to_analyze = wine_data["alcohol"]
print(f"\n--- Thống kê mô tả cho cột '{column_to_analyze.name}' ---")
```

```python
data_mean = np.mean(column_to_analyze)
print(f"Giá trị trung bình (Mean) của '{column_to_analyze.name}': {data_mean:.2f}")
```

```python
data_median = np.median(column_to_analyze)
print(f"Giá trị trung vị (Median) của '{column_to_analyze.name}': {data_median:.2f}")
```

```python
data_mode_result = stats.mode(column_to_analyze, keepdims=True) # keepdims=True để tránh cảnh báo
print(f"Giá trị yếu vị (Mode) của '{column_to_analyze.name}': {data_mode_result.mode[0]:.2f} (xuất hiện {data_mode_result.count[0]} lần)")
```

```python
data_variance = np.var(column_to_analyze)
print(f"Phương sai (Variance) của '{column_to_analyze.name}': {data_variance:.2f}")
```

```python
data_sd = np.std(column_to_analyze)
print(f"Độ lệch chuẩn (Standard Deviation) của '{column_to_analyze.name}': {data_sd:.2f}")
```

```python
data_max = np.max(column_to_analyze)
data_min = np.min(column_to_analyze)
print(f"Giá trị lớn nhất (Max) của '{column_to_analyze.name}': {data_max:.2f}")
print(f"Giá trị nhỏ nhất (Min) của '{column_to_analyze.name}': {data_min:.2f}")
```

```python
data_range = data_max - data_min
print(f"Khoảng biến thiên (Range) của '{column_to_analyze.name}': {data_range:.2f}")
```

```python
data_percentile_60 = np.percentile(column_to_analyze, 60)
print(f"Phân vị thứ 60 (60th Percentile) của '{column_to_analyze.name}': {data_percentile_60:.2f}")
```

```python
data_quartile_75 = np.quantile(column_to_analyze, 0.75)
print(f"Tứ phân vị thứ 3 (Q3 - 75th Percentile) của '{column_to_analyze.name}': {data_quartile_75:.2f}")
```

```python
data_IQR = stats.iqr(column_to_analyze)
print(f"Khoảng Tứ phân vị (IQR) của '{column_to_analyze.name}': {data_IQR:.2f}")
```
